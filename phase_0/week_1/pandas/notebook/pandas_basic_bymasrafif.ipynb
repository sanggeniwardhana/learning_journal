```python
# creating new column based on existing information using loop

# siapkan list kosong
# buat program loop -> selisih pts dengan opp_pts per baris
# simpan hasil per baris ke list kosong
# convert list yang sudah terisi menjadi kolom

hasil = []

# df_new.iloc[:, 10] - kolom pts
# df_new.iloc[:, 14] - kolom opp_pts

for i in range(len(df_new)):
    selisih = df_new.iloc[i, 10] - df_new.iloc[i, 14] # pts - opp_pts
    hasil.append(selisih)

df_new['menggunakan_loop'] = hasil
```

```python
hasil = []

# df_new.iloc[:, 10] - kolom pts
# df_new.iloc[:, 14] - kolom opp_pts

for i in range(len(df_new)):
    selisih = df_new.loc[i, 'pts'] - df_new.loc[i, 'opp_pts'] # pts - opp_pts
    hasil.append(selisih)

df_new['menggunakan_loop_loc'] = hasil
```

```python
# siapin list kosong
# buat program for loop - mengecek apakah menang/kalah/seri
# hasilnya di-append kedalam list
# list di convert menjadi kolom

kesimpulan = []

for i in df_new['menggunakan_loop']:
    if i > 0:
        var_i = 'menang'
    elif i < 0:
        var_i = 'kalah'
    else:
        var_i = 'imbang'
    kesimpulan.append(var_i)

df_new['kesimpulan'] = kesimpulan
```

```python
def pengecekan(data):
    if data > 0:
        var_i = 'menang'
    elif data < 0:
        var_i = 'kalah'
    else:
        var_i = 'imbang'
    return var_i
    
```

```python
kesimpulan = []

for i in df_new['menggunakan_loop']:
    hasil = pengecekan(data=i)
    kesimpulan.append(hasil)

df_new['kesimpulan_cara_2'] = kesimpulan
```
